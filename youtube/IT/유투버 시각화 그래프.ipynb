# 정규화

```python
## Normalization (정규화) : MinMaxScaler() 

from sklearn import preprocessing         #sklearn 파이썬의 머신러닝 모듈

min_max_scaler = preprocessing.MinMaxScaler()

col1=['조회수평균','댓글수평균','좋아요평균','싫어요평균','영상길이평균','구독자수']
x=youtuber[col1].values
x


#정규화시키기        
x_scaled=min_max_scaler.fit_transform(x.astype(float))               #fit_transform 은 실수형태만 받는다. 
x_scaled
```

```python
#정규화 시킨 후의 데이터프레임
youtuber2=pd.DataFrame(x_scaled,columns=col1,index=youtuber.index)
youtuber2
#youtuber2['구독자수']=youtuber['구독자수']
#youtuber2=youtuber2.round(3)
```

```python
## 지금까지 내용을 파일로 저장
youtuber2.to_csv('푸드유투버정규화.csv',encoding='utf-8',
                      sep=',')
```

```python
sns.pairplot(youtuber2,vars=['조회수평균','좋아요평균','댓글수평균'],size=4,kind='reg')   # kind='reg' -> 회귀선
```

```python
## 구독자수와 조회수, 좋아요수와 댓글수에 대한 상관관계

sns.pairplot(youtuber2,x_vars=['구독자수','조회수평균'],
            y_vars=['댓글수평균','좋아요평균'])
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(youtuber2[['조회수평균','댓글수평균','좋아요평균','싫어요평균','영상길이평균']],
           annot=True,fmt='f',cmap='RdPu',linewidths=5)             #annot=     fmt=숫자표시     camp =색     linewidths=테두리 
plt.title('유트브상관관계')
```
